```python
first_df = df[df['day'] <= datetime.strptime('3/8/2021', '%m/%d/%Y')]
first_df = first_df[first_df['day'] >= datetime.strptime('1/1/2021', '%m/%d/%Y')]
first_df = first_df.set_index('day')
first_df
```

```python
second_df = df[df['day'] <= datetime.strptime('11/30/2021', '%m/%d/%Y')]
second_df = second_df[second_df['day'] >= datetime.strptime('3/9/2021', '%m/%d/%Y')]
second_df = second_df.set_index('day')
second_df
```

```python
third_df = df[df['day'] >= datetime.strptime('12/1/2021', '%m/%d/%Y')]
third_df = third_df.set_index('day')
third_df
```
